```python
import pandas as pd
import requests # The requests library 
import bs4 as bs # BeautifulSoup4 is a Python library 
from collections import Counter
import re

df = pd.read_csv('data/clean-airbnb.csv',sep='\t')
df
```

# Price 
By using webscraping, we get price information from airbnb url

```python
# get price information from airbnb url (webscraping)
Price = list()
for i in df['Listing Url']:
    source = requests.get(i) 
    soup = bs.BeautifulSoup(source.text)
    p = soup.find(attrs={'name':'description'})
    while p == None:
        source = requests.get(i) 
        soup = bs.BeautifulSoup(source.text)
        p = soup.find(attrs={'name':'description'})
    p = p['content'].split()[7]
    p = p.replace('.',"")
    p = p.replace('$',"")
    Price.append(p)
    
df['Price']= Price # add price array in dataframe 
```

```python
df = df.loc[df['Price']!='nearby']
```
